Метод выше не дает заменить значения на int. Но удалять строки с отсутствующим годом релиза не стоит, так как статистика по жанру нам понадобится. Заменим пропуски на 0. А лучше на 1900 чтобы на гистограммах небыло большой пропасти в дальнейшем:

`critic_score` оставляем без изменений. Потому как заполнить отсутствующие значения нам нечем. И удалить половину ДФ мы не можем. Тип данных float, работать с данными сможем. Что не сказать про `user_score`

`user_score` - ситуация подобная, выдумать значения мы не можем. Заменить средним тоже не можем. Работаем с тем что есть. Только вот проблема - тип данных `object` Заменить на `float` нам не даст значение `tbd` . В интернете нашел определние на сайте [bf.ru](https://fb.ru/article/171345/tbd---chto-znachit-eto-sokraschenie) означает:
- To Be Determined, то есть "Будет определено". Очень часто подобное сокращение можно встретить в анонсах компьютерных игр. Когда разработчик той или иной игры уже сообщил о том, что она готовится к выходу, но пока еще не может назвать точные данные

По аналогии будем считать, что `TBD` значит что данных попросту еще нет или еще не определены. Выдумать мы их не можем. поэтому заменим `tbd` на пустые значения.

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = np.nan 
```

```python
# посчитаем сколько теперь пропусков в `user_score` 
data['user_score'].isna().sum() / data.shape[0] * 100
```

Стало 54%, а было [считали выше](#nan_count) 40%.Все изза значения `tbd` Зато теперь можем поменять данные на `float`

```python
# Присваиваем тип данных float
data['user_score'] = data['user_score'].astype(float)
```

- Построим график того, как менялись продажи по всем платформам с годами.

```python
for i in data['platform'].unique():
    data[(data['platform'] == i) &
     (data['year_of_release'] != 1900)].groupby('year_of_release')['sum_sales'].sum().plot(figsize=(10,4)).legend(i)
```

Не понимаю зачем я это составил, не зная как сделать нормально=) Не выяснил как посмотреть легенду чтобы оценить продолжительность в годах начала релизов платформы и конец.

Вероятно можно было сделать проще - прошу подсказать) Как Я решил задачу:

```python
platform_list = [] # Создаю пуслой список для названий платформ
platform_duration_yers = [] # Создаю пуслой список для вычисления кол-ва лет

# Создадим цикл, и методом append добавляем постепенно разницу max() and min()
for i in data['platform'].unique():
    
    platform_list.append(i)
    
    platform_duration_yers.append(
        data[(data['platform'] == i) & (data['year_of_release'] != 1900)]['year_of_release'].max()\
        - data[(data['platform'] == i) & (data['year_of_release'] != 1900)]['year_of_release'].min()
    )

# Создаем датафрейм и сразу гистограмму    
pd.DataFrame({'platform':platform_list, 'years':platform_duration_yers})\
    .sort_values('years', ascending=False).reset_index(drop=True)\
    .plot(x='platform', kind='bar', figsize=(10,4));

```

Игры на **pc** выпускаются более 30 лет. Популярные **x360, ps2, psp, wii** около 10 лет.

### Шаг 5. Проверка гипотезы <a id='6'></a>

```python
data['platform'].unique()
```

- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
    - H0: cредний пользовательский рейтинг платформ Xbox One и PC одинаковый
    - H1: cредний пользовательский рейтинг платформ Xbox One и PC различается

```python
alpha = .05

xone = data[(data['platform'] == 'xone') & (data['user_score'] > 0)]['user_score']   #  & (data['user_score'] > 0)
pc = data[(data['platform'] == 'pc') & (data['user_score'] > 0)]['user_score']

results = st.ttest_ind(xone, pc)
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

- Средние пользовательские рейтинги жанров Action и Sports разные.
    - H0: cредний пользовательский рейтинг жанров Action и Sports одинаковые
    - H1: cредний пользовательский рейтинг жанров Action и Sports разные

```python
action = data[(data['genre'] == 'action') & (data['user_score'] > 0)]['user_score']
sports = data[(data['genre'] == 'sports') & (data['user_score'] > 0)]['user_score']

results = st.ttest_ind(action, sports)
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

    - В первом случае нулевая гипотека выдвинута о схожести, поэтому при p < 0.05 мы можем сделать вывод что данные раздичаются
    - Во втором случае нулевая гпотеза выдвинута о схожести, поэтому при полученном p **<** 0.05 мы можем сделать вывод что данные схожи.А значит не отвергаем нулевую гипотезу.

```python
# Приведены два датасета: сумма покупок, совершённых за месяц посетителями ...

sample_1 = [3071, 3636, 3454, 3151, 2185, 3259, 1727, 2263, 2015,
2582, 4815, 633, 3186, 887, 2028, 3589, 2564, 1422, 1785,
3180, 1770, 2716, 2546, 1848, 4644, 3134, 475, 2686,
1838, 3352]
sample_2 = [1211, 1228, 2157, 3699, 600, 1898, 1688, 1420, 5048, 3007,
509, 3777, 5583, 3949, 121, 1674, 4300, 1338, 3066,
3562, 1010, 2311, 462, 863, 2021, 528, 1849, 255,
1740, 2596]
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(
sample_1,
sample_2)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:

    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
# check
# круги + категория другие
def graph (df, year, region, name, axes):
    
    df = df.query('year_of_release >= @year')
    
    sales = df.pivot_table(index='platform', 
                           values=region, 
                           aggfunc='sum').nlargest(5, region)
    
    sales = sales.reset_index()
    
    sales = (
            sales.append({'platform': 'Other', region: df[region].sum() 
                       - sales[region].sum()}, ignore_index= True)
         )
    
    
    sales.columns = ['platform', 'sales']
      
    labels_c=sales.platform
    colours = {'Wii':'C0', 'NES':'C1', 'GB':'C2', 'DS':'C3', 'X360':'C4', 
    'PS3':'C5', 'PS2':'C6', 'SNES':'C7', 'GBA':'C8',
               'PS4':'steelblue', '3DS':'orange', 
               'N64':'C11', 'PS':'C12', 'XB':'C13', 'PC':'C14', '2600':'C15', 'PSP':'C16', 
               'XOne':'C17',
               'WiiU':'C18', 'GC':'C19', 'GEN':'C20', 'DC':'C21', 'PSV':'C22', 
               'SAT':'C23', 'SCD':'C24', 'WS':'C25', 'NG':'C26', 
               'TG16':'C27', '3DO':'C28', 'GG':'C29', 'PCFX':'C30', 'Other':'darkred'}
   
    sales.plot(kind='pie',
               y="sales",
               
               autopct='%1.0f%%',
               wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
               textprops={'size': 'x-large'}, 
               labels= labels_c,
               colors=[colours[key] for key in labels_c],
               legend=False, 
               title = f"Популярность платформ в {name} ", 
               ax = axes).set(ylabel='')
    
    plt.tight_layout()
```

```python
# check
# круги в ряд
fig, axes = plt.subplots(1, 3, figsize = (15,6))
fig.suptitle('Обзор рынка платформ (портрет покупателя)', fontsize = 30, fontweight='bold')

x_year = 2011

graph(data_games, x_year, 'na_sales', 'Северной Америке', axes[0])
graph(data_games, x_year,'eu_sales', 'Европе', axes[1])
graph(data_games, x_year, 'jp_sales', 'Японии', axes[2])
```
