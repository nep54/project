#### –í—ã–≤–µ–¥–µ–º –ø–µ—Ä–≤—ã–µ —Å—Ç—Ä–æ–∫–∏ —Ç–∞–±–ª–∏—Ü

```python
# –£–ø—Ä–æ—Å—Ç–∏–º, –Ω–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ –∑–∞–ø—Ä–æ—Å–æ–≤ SQL
def sql(query):
    return pd.io.sql.read_sql(query, con = engine) 

def info_all(data):
    '''–§—É–Ω–∫—Ü–∏—è –≤—ã–≤–∞–µ–¥–µ—Ç –≤—Å–µ —Ä—è–¥–æ–≤—ã–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –ø—Ä–∏ –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–µ –¥–∞–Ω–Ω—ã—Ö'''
    display(data.head(5))
    print()
    print(data.info())
    print()
    print(f'–ö–æ–ª-–≤–æ –¥—É–±–ª–µ–π {data.duplicated().sum()} —Å—Ç—Ä–æ–∫')
    print()
    print(f'–ö–æ–ª-–≤–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤')
    print(data.isna().sum())
    print()
    display(data.describe())
    print('-----------------------------------------------------------------------------------------')
```

### –ó–∞–¥–∞–Ω–∏–µ 2: –î–ª—è –∫–∞–∂–¥–æ–π –∫–Ω–∏–≥–∏ –ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±–∑–æ—Ä–æ–≤ –∏ —Å—Ä–µ–¥–Ω—é—é –æ—Ü–µ–Ω–∫—É;

```python
reviews_rating = sql("""

SELECT 
    books.title,
    AVG(ratings.rating),
    COUNT(DISTINCT reviews.text)
FROM books 
LEFT JOIN ratings ON ratings.book_id = books.book_id
LEFT JOIN reviews ON reviews.book_id = books.book_id
GROUP BY
    books.title

""")
reviews_rating
```

### –ó–∞–¥–∞–Ω–∏–µ 3: –û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –∏–∑–¥–∞—Ç–µ–ª—å—Å—Ç–≤–æ, –∫–æ—Ç–æ—Ä–æ–µ –≤—ã–ø—É—Å—Ç–∏–ª–æ –Ω–∞–∏–±–æ–ª—å—à–µ–µ —á–∏—Å–ª–æ –∫–Ω–∏–≥ —Ç–æ–ª—â–µ 50 —Å—Ç—Ä–∞–Ω–∏—Ü
- —Ç–∞–∫ –º—ã –∏—Å–∫–ª—é—á–∏–º –∏–∑ –∞–Ω–∞–ª–∏–∑–∞ –±—Ä–æ—à—é—Ä—ã;

```python
sql("""

SELECT 
    p.publisher,
    COUNT(b.title) AS count
FROM 
    books AS b
LEFT JOIN 
    publishers AS p ON p.publisher_id = b.publisher_id
WHERE 
    b.num_pages > 50
GROUP BY 
    p.publisher
ORDER BY 
    count DESC
LIMIT (1)

""")
```

### –ó–∞–¥–∞–Ω–∏–µ 4: –û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ –∞–≤—Ç–æ—Ä–∞ —Å —Å–∞–º–æ–π –≤—ã—Å–æ–∫–æ–π —Å—Ä–µ–¥–Ω–µ–π –æ—Ü–µ–Ω–∫–æ–π –∫–Ω–∏–≥
- –£—á–∏—Ç—ã–≤–∞–µ–º —Ç–æ–ª—å–∫–æ –∫–Ω–∏–≥–∏ —Å 50 –∏ –±–æ–ª–µ–µ –æ—Ü–µ–Ω–∫–∞–º–∏;

```python
sql('''
SELECT
    y.author,
    AVG(y.avg_rat) as avg_rat
FROM
    (SELECT 
        b.title,
        a.author,
        AVG(r.rating) AS avg_rat
    FROM 
        books AS b
    LEFT JOIN 
        authors AS a ON a.author_id = b.author_id
    LEFT JOIN 
        ratings AS r ON r.book_id = b.book_id
    GROUP BY 
        b.title,
        a.author
    HAVING    
        COUNT(r.rating) >= 50) AS y
GROUP BY 
    y.author
ORDER BY
    avg_rat DESC
LIMIT (1)  

''')
```

### –ó–∞–¥–∞–Ω–∏–µ 5: –ü–æ—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–±–∑–æ—Ä–æ–≤ –æ—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ—Å—Ç–∞–≤–∏–ª–∏ –±–æ–ª—å—à–µ 50 –æ—Ü–µ–Ω–æ–∫.

```python
sql('''
SELECT ROUND(AVG(y.count),2)
FROM
    (SELECT 
        r.username,
        COUNT(r.text)

    FROM
        (SELECT 
            username,
            COUNT(rating_id) as count_rat
        FROM 
            ratings
        GROUP BY
            username
        HAVING COUNT(rating_id) > 50) AS count_rat
    INNER JOIN
        reviews AS r ON r.username = count_rat.username

    GROUP BY 
        r.username) as y
''')
```
