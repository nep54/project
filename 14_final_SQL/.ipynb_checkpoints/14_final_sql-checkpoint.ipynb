### Задание 2: Для каждой книги посчитаем количество обзоров и среднюю оценку;

```python
reviews_rating = sql("""

SELECT 
    books.title,
    AVG(ratings.rating),
    COUNT(DISTINCT reviews.text)
FROM books 
LEFT JOIN ratings ON ratings.book_id = books.book_id
LEFT JOIN reviews ON reviews.book_id = books.book_id
GROUP BY
    books.title

""")
reviews_rating
```

### Задание 3: Определить издательство, которое выпустило наибольшее число книг толще 50 страниц
- так мы исключим из анализа брошюры;

```python
sql("""

SELECT 
    p.publisher,
    COUNT(b.title) AS count
FROM 
    books AS b
LEFT JOIN 
    publishers AS p ON p.publisher_id = b.publisher_id
WHERE 
    b.num_pages > 50
GROUP BY 
    p.publisher
ORDER BY 
    count DESC
LIMIT (1)

""")
```

- Наибольшее кол-во книг (толще 50 страниц) выпустило издательство `Penguin Books`, в колличестве 42 штук. 

### Задание 4: Определите автора с самой высокой средней оценкой книг
- Учитываем только книги с 50 и более оценками;

```python
sql('''
SELECT
    y.author,
    AVG(y.avg_rat) as avg_rat
FROM
    (SELECT 
        b.title,
        a.author,
        AVG(r.rating) AS avg_rat
    FROM 
        books AS b
    LEFT JOIN 
        authors AS a ON a.author_id = b.author_id
    LEFT JOIN 
        ratings AS r ON r.book_id = b.book_id
    GROUP BY 
        b.title,
        a.author
    HAVING    
        COUNT(r.rating) >= 50) AS y
GROUP BY 
    y.author
ORDER BY
    avg_rat DESC
LIMIT (1)  

''')
```

- Самая высока средняя оценка у автора `J.K. Rowling/Mary GrandPré` с рейтингом `4.28` 

### Задание 5: Посчитаем среднее количество обзоров от пользователей, которые поставили больше 50 оценок.

```python
sql('''
SELECT ROUND(AVG(y.count),2)
FROM
    (SELECT 
        r.username,
        COUNT(r.text)

    FROM
        (SELECT 
            username,
            COUNT(rating_id) as count_rat
        FROM 
            ratings
        GROUP BY
            username
        HAVING COUNT(rating_id) > 50) AS count_rat
    INNER JOIN
        reviews AS r ON r.username = count_rat.username

    GROUP BY 
        r.username) as y
''')
```

- 24 обзора в среднем написали пользователи, которые оставили более 50 оценок
